```python
# Plot distribution of labels
from src.data_loading import load_dataframe_labels


labels_train = load_dataframe_labels("data_train")

# labels_train is a df with 4 columns : ID,HOME_WINS,DRAW,AWAY_WINS
# We want to plot the distribution of the labels
labels_train = labels_train[["HOME_WINS", "DRAW", "AWAY_WINS"]]
labels_train = labels_train.idxmax(axis=1)
labels_train.value_counts().plot(kind='bar')
plt.title("Distribution of labels")
plt.show()
```

```python
# Plot distribution of predictions
pred_train = pd.read_csv("data/predictions_train/fold0.csv")
pred_train = pred_train[["HOME_WINS", "DRAW", "AWAY_WINS"]]
pred_train = pred_train.idxmax(axis=1)
pred_train.value_counts().plot(kind='bar')
plt.title("Distribution of predictions for train set")
plt.show()
```
